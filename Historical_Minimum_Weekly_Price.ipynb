Hystorical Behavior of the Selected stock
===========================

```python

### GENERIC INFORMATION ABOUT stock and its HISTORICAL BEHAVIOR
# Time Informations
day, month, year = stock_utils.month_year()
stock_under_test = input("Enter the stock ticker symbol: ")
#stock_under_test = "BTC-EUR"
category = input("Insert Category ETF, Crypto, Stock")
#category = "Crypto"
#start_date = input("Enter the start date (YYYY-MM-DD): ")
start_date = "2018-01-01"
end_date = f"{year}-{month}-{day}"
stock_data_object = stock_utils.get_stock_data(stock_under_test)
stock_data_values = stock_utils.get_stock_with_date_index_data(
        stock_data_object, category, start_date, end_date, ma_period=200
        )

# Assicurati che l'indice sia di tipo DatetimeIndex
stock_data_values.index = pd.to_datetime(stock_data_values.index)

# Aggiungi una colonna per il giorno della settimana
stock_data_values['weekday'] = stock_data_values.index.day_name()

# Raggruppa per settimana e trova il giorno con il prezzo più basso per ogni settimana
weekly_min = stock_data_values.groupby(pd.Grouper(freq='W'))['stock_price'].idxmin()
weekly_max = stock_data_values.groupby(pd.Grouper(freq='W'))['stock_price'].idxmax()

print(f"weekly_min: {weekly_min}")
print(f"weekly_max: {weekly_max}")

# Conta quale giorno della settimana ha avuto il prezzo più basso più frequentemente
weekday_counts_min = stock_data_values.loc[weekly_min, 'weekday'].value_counts()
weekday_counts_max = stock_data_values.loc[weekly_max, 'weekday'].value_counts()

# Stampa il conteggio di ciascun giorno della settimana
print(f"weekday_counts : {weekday_counts_min}")
print(f"weekday_counts : {weekday_counts_max}")

# Stampa il giorno della settimana che ha avuto il prezzo più basso più frequentemente
most_frequent_min_day = weekday_counts_min.idxmax()
most_frequent_max_day = weekday_counts_max.idxmax()
print(f"Il giorno della settimana che ha esibito più volte il valore più basso della settimana è: {most_frequent_min_day}")
print(f"Il giorno della settimana che ha esibito più volte il valore più alto della settimana è: {most_frequent_max_day}")
```
